```python
import matplotlib.pyplot as plt

# 라벨별로 데이터 필터링
label_data = {}

for label in df['Label'].unique():
    label_data[label] = df[df['Label'] == label]

# 선 그래프 시각화
fig, axes = plt.subplots(len(label_data), 1, figsize=(12, 6*len(label_data)), sharex=True)

for i, (label, data) in enumerate(label_data.items()):
    axes[i].plot(data.index, data['GyrX'], label='GyrX')
    axes[i].plot(data.index, data['GyrY'], label='GyrY')
    axes[i].plot(data.index, data['GyrZ'], label='GyrZ')
    axes[i].set_ylabel('Gyro Data')
    axes[i].set_title(label)
    axes[i].legend()

axes[-1].set_xlabel('Index')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# TSNE를 위한 데이터 준비
features = df[['AccX', 'AccY', 'AccZ', 'GyrX', 'GyrY', 'GyrZ']]
labels = df['Label']

# TSNE 모델 학습
tsne = TSNE(n_components=2, random_state=42)
embedded_features = tsne.fit_transform(features)

# 시각화
plt.figure(figsize=(10, 8))
scatter = plt.scatter(embedded_features[:, 0], embedded_features[:, 1], c=labels.cat.codes, cmap='rainbow')
plt.legend(handles=scatter.legend_elements()[0], labels=labels.unique(), title='Label')
plt.xlabel('TSNE Dimension 1')
plt.ylabel('TSNE Dimension 2')
plt.title('TSNE Visualization')
plt.show()
```
