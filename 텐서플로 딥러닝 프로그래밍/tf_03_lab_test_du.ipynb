````python
# 필요한 라이브러리 import
import numpy as np
import pandas as pd
import tensorflow as tf

# 데이터 로드
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']

raw_dataset = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)

# 결측값 처리
dataset = raw_dataset.fillna(raw_dataset.mean())

# 데이터 분석
mpg_sum = dataset['MPG'].sum()
mpg_mean = dataset['MPG'].mean()
mpg_min = dataset['MPG'].min()
mpg_max = dataset['MPG'].max()

cylinders_sum = dataset['Cylinders'].sum()
cylinders_mean = dataset['Cylinders'].mean()
cylinders_min = dataset['Cylinders'].min()
cylinders_max = dataset['Cylinders'].max()

horsepower_sum = dataset['Horsepower'].sum()
horsepower_mean = dataset['Horsepower'].mean()
horsepower_min = dataset['Horsepower'].min()
horsepower_max = dataset['Horsepower'].max()

weight_sum = dataset['Weight'].sum()
weight_mean = dataset['Weight'].mean()
weight_min = dataset['Weight'].min()
weight_max = dataset['Weight'].max()

largest_hp_idx = dataset['Horsepower'].idxmax()
largest_hp = dataset.loc[largest_hp_idx]

# 결과 출력
print("MPG 합계: ", mpg_sum)
print("MPG 평균: ", mpg_mean)
print("MPG 최솟값: ", mpg_min)
print("MPG 최댓값: ", mpg_max)

print("Cylinders 합계: ", cylinders_sum)
print("Cylinders 평균: ", cylinders_mean)
print("Cylinders 최솟값: ", cylinders_min)
print("Cylinders 최댓값: ", cylinders_max)

print("Horsepower 합계: ", horsepower_sum)
print("Horsepower 평균: ", horsepower_mean)
print("Horsepower 최솟값: ", horsepower_min)
print("Horsepower 최댓값: ", horsepower_max)

print("Weight 합계: ", weight_sum)
print("Weight 평균: ", weight_mean)
print("Weight 최솟값: ", weight_min)
print("Weight 최댓값: ", weight_max)

print("Horsepower가 가장 큰 데이터의 row 번호: ", largest_hp_idx)
print("Horsepower가 가장 큰 데이터: ", largest_hp)
````

````python
import pathlib

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
````

````python
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
````

````python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
raw_dataset = pd.read_csv(dataset_path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)

dataset = raw_dataset.copy()
dataset.tail()
````

````python
dataset.isna().sum()
````

````python
dataset = dataset.dropna()
````

````python
origin = dataset.pop('Origin')
````

````python
dataset['USA'] = (origin == 1)*1.0
dataset['Europe'] = (origin == 2)*1.0
dataset['Japan'] = (origin == 3)*1.0
dataset.tail()
````

````python
train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)
````

````python
sns.pairplot(train_dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind="kde")
````

````python
train_stats = train_dataset.describe()
train_stats.pop("MPG")
train_stats = train_stats.transpose()
train_stats
````

````python
train_labels = train_dataset.pop('MPG')
test_labels = test_dataset.pop('MPG')
````

````python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
````

````python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
````

````python
model = build_model()
````

````python
model.summary()
````

````python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
````

````python
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
````

````python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
````

````python
import matplotlib.pyplot as plt

def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,5])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,20])
  plt.legend()
  plt.show()

plot_history(history)
````

````python
model = build_model()

early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

plot_history(history)
````

````python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)

print("테스트 세트의 평균 절대 오차: {:5.2f} MPG".format(mae))
````

````python
test_predictions = model.predict(normed_test_data).flatten()

plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])
````

````python
error = test_predictions - test_labels
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]")
_ = plt.ylabel("Count")
````
